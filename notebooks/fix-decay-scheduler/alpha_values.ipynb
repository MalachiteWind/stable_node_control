```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
Ti = np.linspace(0,10,1001)
gamma = 0.5
steps = 6
alphas = [1*(gamma**i) for i in range(steps)]
alphas = np.concatenate([alphas, [0]])
```

```python
alphas = np.concatenate([alphas, [0]])
```

```python
plt.figure(figsize=(15,5))
for a in alphas:
    label = rf"$\alpha$ = {a:.1e}"
    plt.plot(Ti, np.exp(-a*Ti), label=label, lw=4)
plt.legend(prop={'family':'serif'})
plt.title("Decay rates", family='serif',size=30)
plt.xlabel(r"$t$",size=30)
plt.ylabel(r"$w(t)$", size=30)
plt.show()
```
